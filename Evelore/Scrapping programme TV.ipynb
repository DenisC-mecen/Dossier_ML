### Chemin

```python
def chemin(Dates):
    url = "http://programme-tv.nouvelobs.com/programme-tv/"
    Chemins = []
    for i in range(len(Dates)):
        Chemin = url + str(Dates[i]) + "/"
        Chemins.append(Chemin)
    return Chemins
```
