```python
Dates = url.dates(2019)
```

```python
Chemins = url.chemin(Dates)
len(Chemins)
```

```python
for i in range(len(Chemins)):
    arbre = fct.tree(Chemins[i])
    V4 = fct.jour(arbre)
    V5 = fct.chaines(arbre)
    V6 = fct.heures1(arbre)
    V7 = fct.titres1(arbre)
    V8 = fct.type_prgm1(arbre)
    V9 = fct.temps_prgm1(arbre)
    V10 = fct.nbre_ep1(arbre)
    V11 = fct.ages1(arbre)
    V12 = fct.heures2(arbre)
    V13 = fct.titres2(arbre)
    V14 = fct.type_prgm2(arbre)
    V15 = fct.temps_prgm2(arbre)
    V16 = fct.nbre_ep2(arbre)
    V17 = fct.ages2(arbre)
    Donnees = gen.Donnee(V4,V5,V6,V7,V8,V9,V10,V11,V12,V13,V14,V15,V16,V17,Donnees)
```
