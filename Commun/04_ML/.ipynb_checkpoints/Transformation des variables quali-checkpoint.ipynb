# Recodage par label encoder

```python
from sklearn.preprocessing import (OneHotEncoder,
                                   LabelEncoder,
                                   StandardScaler,
                                   Normalizer,
                                   MinMaxScaler,
                                  )
```

```python
gestion_jour = LabelEncoder()
gestion_mois = LabelEncoder()
```

```python
Jour_lab = gestion_jour.fit_transform(data["Jour"].astype(str))
Mois_lab = gestion_mois.fit_transform(data["Mois"].astype(str))
```

```python
gestion_heure_prgm1 = LabelEncoder()
gestion_heure_prgm2 = LabelEncoder()
```

```python
Heure_prgm1_lab = gestion_heure_prgm1.fit_transform(data["Heure_prgm1"].astype(str))
Heure_prgm2_lab = gestion_heure_prgm2.fit_transform(data["Heure_prgm2"].astype(str))
```

```python
gestion_age_prgm1 = LabelEncoder()
gestion_age_prgm2 = LabelEncoder()
```

```python
Age_conseille_prgm1_lab = gestion_age_prgm1.fit_transform(data["Age_conseille_prgm1"].astype(str))
Age_conseille_prgm2_lab = gestion_age_prgm2.fit_transform(data["Age_conseille_prgm2"].astype(str))
```

# Recodage par One Out Encoder

```python
gestion_chaine = OneHotEncoder()
```

```python
Chaine_ohe = gestion_chaine.fit_transform(data["Chaine"]
                                  .astype(str)
                                  .values
                                  .reshape(-1, 1)
                                 )
```

```python
gestion_type_prgm1 = OneHotEncoder()
gestion_type_prgm2 = OneHotEncoder()
```

```python
Type_prgm1_ohe = gestion_type_prgm1.fit_transform(data["Type_prgm1"]
                                  .astype(str)
                                  .values
                                  .reshape(-1, 1)
                                 )
Type_prgm2_ohe = gestion_type_prgm2.fit_transform(data["Type_prgm2"]
                                  .astype(str)
                                  .values
                                  .reshape(-1, 1)
                                 )
```
