```python
#Vecteur des dates
Dates = []
for k in range(6):
    for j in range(12):
        for i in range(31):
            'Jour'
            if i < 9:
                JJ = "0" + str(i+1)
            else:
                JJ = str(i+1)
            'Mois'
            if j < 9:
                MM = "0" + str(j+1)
            else:
                MM = str(j+1)
            'Année'
            AAAA = k + 2014
            AAAA = str(AAAA)
            'Date'
            Date = JJ + "-" + MM + "-" + AAAA
            Dates.append(Date)
            
```

## Conversion de logo chaine en chaine

```python
def recodage(col, codeDict):
    colCoded = pd.Series(col, copy=True)
    for key, value in codeDict.items():
        colCoded.replace(key, value, inplace=True)
    return colCoded
```
